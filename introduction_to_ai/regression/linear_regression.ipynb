```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 5.0]

def forward(x):
  return x * w

def loss(y_pred, y):
  return (y_pred - y) * (y_pred-y)

w = 0.00

for epoch in range(30):
  for x,y in zip(x_data,y_data):
    y_pred = forward(x)
    l = loss(y_pred, y)
    # if x==3 :
    #   print("Epoch", epoch, "\t(x,y):", x, y, "\tw=", "$.2f"%w, "\tloss=", "%.2f"%l)
    w = w + 0.05
  plt.scatter(w, l, color = 'r')
```

```python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 5.0]
learning_rate = 0.005

def gradient(x,y):
  return 2 * x * (x * w - y)

w = 1.00
for epoch in range(30):
  for x,y in zip(x_data,y_data):
    y_pred = forward(x)
    l = loss(y_pred, y)
    grad = gradient(x, y)

    w = w-learning_rate*grad
  
  plt.scatter(w, l, color = 'r')
```

```python
x_data = [1.0, 2.0, 3.0, 4.0, 6.0]
y_data = [2.0, 4.0, 6.0, 8.0, 12.0]
w = 0.0
b = 0.0

n_data = len(x_data)
epochs = 100
learning_rate = 0.01
for i in range(epochs):
  print("Process ", i+1)
  for x_i, y_i in zip(x_data, y_data):
    y_hat = x_i*w+b
    loss = ((y_hat-y_i)**2) / n_data
    grad_w = ((w*x_i-y_i+b)*2*x_i)/n_data
    grad_b = ((w*x_i-y_i+b)*2)/n_data
    w -= learning_rate * grad_w
    b -= learning_rate * grad_b
    print("weight = ", w, "\t\t", "bias= ", b)
  plt.scatter(w,loss, color ='r')
```
